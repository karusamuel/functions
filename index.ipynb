## Introduction to Functions
What is a Function?

A function is a reusable block of code designed to perform a specific task.
Instead of writing the same code multiple times, we write it once inside a function and call it whenever needed.

Think of a function like a machine:

Input → Processing → Output

#### Why Use Functions?

Functions help us:

✅ Avoid repeating code

✅ Make programs easier to read and maintain

✅ Break large problems into smaller, manageable parts

✅ Write cleaner and more organized code

#### Defining vs Executing a Function

There are two steps when working with functions:

    - Defining (Declaring) a function

    - Executing (Calling) a function

Defining happens first:

````python
# example
# define a function
def greet():
    print("hello")
    print("my name is sam")
    print("am new to python")
    print("Hello world")
````

````python
# example 
# call a function

greet()
````

#### Returning Values from a Function

By default, functions do not return anything.

```python
def example():
    x = 10
```
To send data out of a function, we use return

````python
# An example of a function with return statement 
A = 5
B = 10
def add():
    print("starting")
    ans = A+B
    print("end")
    return ans
````

````python
answer = add()
answer
````

#### Functions with Arguments

Arguments allow us to pass data into a function, making it more flexible.

````python
# example def
def greet(name):
    print(f"hello {name} nice to meet you")
````

````python
# example call
greet(name="jane")
````

````python
greet(name="sam")
````

````python
greet(name = " a very long name here")
````

````python
def greet(name,time="Morning"):
    if(time=="Morning"):
        print(f"Goodmorning {name} welcome to moringa")
    else:
        print(f"hello {name} welcome to moringa")
````

````python
greet("samuel karu")
````

````python
greet(time="evening",name="samuel karu")
````
