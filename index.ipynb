What is a Function?

A function is a reusable block of code designed to perform a specific task.
Instead of writing the same code multiple times, we write it once inside a function and call it whenever needed.

Think of a function like a machine:

Input → Processing → Output

#### Why Use Functions?

Functions help us:

✅ Avoid repeating code

✅ Make programs easier to read and maintain

✅ Break large problems into smaller, manageable parts

✅ Write cleaner and more organized code

## Files in python 

In Python, a **file** is used to store data permanently on a computer. Unlike variables, which only exist while a program is running, files allow data to be saved and accessed later.

### Why use files?

* To save program output
* To read existing data
* To share data between programs

### Basic file operations

Python works with files using three main modes:

* `r` (read): to read data from a file
* `w` (write): to create a new file or overwrite an existing one
* `a` (append): to add data to an existing file

### Older method (without context manager)

```python
# Writing to a file
file = open("example.txt", "w")
file.write("Hello, world!\n")
file.close()

# Reading from a file
file = open("example.txt", "r")
content = file.read()
print(content)
file.close()
```

````python
# test 
file = open("sam.txt","+w")
file.write("hello my name is jane")
file.close()

````

### Example using `with open()` (recommended)

```python
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, world!\n")

# Reading from the same file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```


````python
# example 

with open("sam.txt","r+") as file:
    print(file.readlines())
````

````python
with open("sam.txt","+a") as file:
    file.write("i work here")
````

````python
with open("sam.txt","") as file:
    file.writelines(["i work here \n","i work here too \n"])
````


### Key difference

* The `with open()` method automatically closes the file
* The older method requires you to manually call `file.close()`
* Forgetting to close a file can lead to bugs or data issues
`use context managers`

## Regular Expressions (RegEx) in Python

A **regular expression (RegEx)** is a pattern used to search, match, or manipulate text. It is useful when you need to find specific patterns in strings, such as emails, phone numbers, or words.

### Why use RegEx?

* Search for patterns in text
* Validate input (e.g., emails, passwords)
* Extract specific parts of a string
* Replace text patterns

### The `re` module

Python provides the built-in `re` module to work with regular expressions.

### Basic Example

```python
import re

text = "My phone number is 0712345678"

# Search for a number pattern
match = re.search(r"\d+", text)

if match:
    print(match.group())  # Output: 0712345678
```

````python
import re 
````

### Common RegEx Patterns

| Pattern | Meaning                                 |
| ------- | --------------------------------------- |
| `\d`    | Any digit (0–9)                         |
| `\w`    | Any word character (letters, digits, _) |
| `\s`    | Any whitespace                          |
| `.`     | Any character except newline            |
| `+`     | One or more occurrences                 |
| `*`     | Zero or more occurrences                |
| `^`     | Start of string                         |
| `$`     | End of string                           |


## Common RegEx Functions in Python (Examples)

### 1. `re.search()` — finds the first match anywhere in the string

```python
import re

text = "My number is 12345"

match = re.search(r"\d+", text)

if match:
    print(match.group())  # Output: 12345
```


### 2. `re.findall()` — returns all matches as a list

```python
import re

text = "I have 2 apples and 3 bananas"

numbers = re.findall(r"\d+", text)

print(numbers)  # Output: ['2', '3']
```


### 3. `re.match()` — checks for a match at the beginning of the string

```python
import re

text = "Hello world"

match = re.match(r"Hello", text)

if match:
    print("Match found at the beginning")
else:
    print("No match")
```

### 4. `re.sub()` — replaces matches with new text

```python
import re

text = "I have 2 apples"

result = re.sub(r"\d+", "many", text)

print(result)  # Output: I have many apples
```
